#### Renaming, Splitting, Dropping, Rearranging of the Columns

```python
#checking the name of the columns
print (df.columns)
```

```python
#rename the column names
df = df.rename (columns = {"Mat" : "Match",
                           "Player": "Player_Name",
                          4: "4s",
                          5: "5s" })
#splitting the "Player" column to get information about "Country"
df [["Player_Name", "Country"]] = df["Player_Name"].str.split ("(", expand = True)

# dropping the "Player" column
# df = df.drop ("Player", axis = 1 )

# remove ")" and "INDIA" as "IND" from the "Country" column
df ["Country"] = df["Country"].str.replace (")", "")
df ["Country"] = df["Country"].str.replace ("INDIA", "IND")

# check the new column name
display (df.head ())
# display (df.tail (10))
```

#### How many players played for ICC
13 players played for ICC

```python
def icc_check (x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df ["Played_for_ICC"] = df ["Country"].apply(icc_check)

display (df.head (10))
```

```python
print (df["Played_for_ICC"].value_counts())
```

#### Span of the players

```python
#split the "Span" column based on the "-"

df [["start_year", "end_year"]] = df ["Span"].str.split ("-", expand = True)

#dropping the "Span" column
df = df.drop ("Span", axis =1)

#convert data types from string to integer
df ["start_year"] = df ["start_year"].astype("int")
df ["end_year"] = df ["end_year"].astype("int")

#years played = (end_year - start_year) + 1
df ["years_played"] = (df ["end_year"]- df ["start_year"]) + 1

display (df.head())
```
